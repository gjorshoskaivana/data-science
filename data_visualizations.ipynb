## Relationship between certain variables

```
# relationship between volatily acidity and citric acid
fig = plt.figure()
select_data.plot.scatter(x= 'volatile acidity', y = 'citric acid', alpha = 0.75,rot=0)
plt.xticks(rotation=90)
```

```
# relationship between volatily acidity and density
fig = plt.figure()
select_data.plot.scatter(x= 'volatile acidity', y = 'density', alpha = 0.75,rot=0)
plt.xticks(rotation=90)
```

```
# relationship between volatily acidity and pH
fig = plt.figure()
select_data.plot.scatter(x= 'volatile acidity', y = 'pH', alpha = 0.75,rot=0)
plt.xticks(rotation=90)
```

```
# relationship between citric acid and density
fig = plt.figure()
select_data.plot.scatter(x= 'citric acid', y = 'density', alpha = 0.75,rot=0)
plt.xticks(rotation=90)
```

```
# relationship between citric acid and pH
fig = plt.figure()
select_data.plot.scatter(x= 'citric acid', y = 'pH', alpha = 0.75,rot=0)
plt.xticks(rotation=90)
```

```
# relationship between density and pH
fig = plt.figure()
select_data.plot.scatter(x= 'density', y = 'pH', alpha = 0.75,rot=0)
plt.xticks(rotation=90)
```

The following visualization shows the quality of wine in relation to the acidity, counted with the pH balance.







```
#add your code
import plotly.express as px

ds = dataset.groupby(['volatile acidity','quality'])['pH'].count().reset_index()
ds.columns = ['volatile acidity', 'quality', 'count pH']
ds
```

```
fig = px.bar(
    ds, 
    x='quality', 
    y="count pH", 
    color = 'volatile acidity',
    orientation='v', 
    title='the quality of wine per pH balance and acidity', 
    width=800,
    height=500
)
fig.show()
```
